```python
import requests as req
from json import loads
from base64 import b64decode
from PyPDF2 import PdfReader, PdfFileWriter
#PdfFileReader
```

# Scraping and convert pdf information to text

```python
# Ruta de descarga
#path_output = "C:/Users/WIN/Scraping_Commerce_Web_Page/2_output/"
path_output ="/content/2_output/"
# Cabecera de configuracion del navegador
header = {"User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/107.0.0.0 Safari/537.36",
          "Host": "recibos.sigcomt.com:9001",
          "Accept-Encoding": "gzip, deflate, br",
          "Content-Type": "application/json"}

# Range of random gas supply
#for i in range(90000,110000):
for i in (list(range(99990,100010)) + list(range(160000,160060))):
    
    datos = {"year": 2022,
             "clientCode": str(i),
             "month": 10}

    # Api from get the information of the gas supply
    resp = req.post("https://recibos.sigcomt.com:9001/Back/api/ReceiptView/ExternalPdf", headers = header, json = datos)

    #print(resp.text)

    # Get the information
    dato = loads(resp.text)['data']

    # If the data doesn't exist go to "else"
    if dato is not None:
        
        #I decode the data from 64 base
        dato_bytes = b64decode(dato, validate=True)

        # If the data is pdf
        if dato_bytes[0:4] != b'%PDF':
            print('Missing the PDF file signature')
        else:
            
            # Save the PDF data of the gas supply
            f = open(path_output + 'recibo_suministro_' + str(i) + '.pdf', 'wb')
            
            print("******* Recibo encontrado del suministro: " + str(i) + " *******")

            f.write(dato_bytes)
            f.close()
            
            # Read the pdf information
            fpdf = PdfReader(path_output + 'recibo_suministro_' + str(i) + '.pdf')
            
            # Information of the pdf
            txtfromfpdf = ""
    
            Customer_kind = "B2C"

            #if(fpdf.numPages>1):
            #len(fpdf.pages)=fpdf.numPages
            if(len(fpdf.pages)>1): 
                Customer_kind = "B2B"

            for page_num in range(len(fpdf.pages)):
                #print("Página " + str(page_num) + " :")

                #pagefpdf = fpdf.getPage(page_num)
                pagefpdf = fpdf.pages[page_num]
                

                try:
                    txtfrompagefpdf = pagefpdf.extract_text()
                    txtfromfpdf = txtfromfpdf + txtfrompagefpdf
                except:
                    pass

            txtfromfpdf = txtfromfpdf.split("\n")
            
            print("------- " + Customer_kind + " -------")
            print("Suministro: " + str(i))

            if(Customer_kind == "B2C"):
                
                
                print("Nombre Completo: " + str(txtfromfpdf[-4]).strip())
                print("Direccion: " + str(txtfromfpdf[-3]).strip() + " " + str(txtfromfpdf[-2]).strip())

            else:
                print("Razon Social: " + str(txtfromfpdf[0]).strip())
                print(str(txtfromfpdf[3]).strip())#3
                print("Direccion de Facturacion: " + str(txtfromfpdf[1]).strip() + " " + str(txtfromfpdf[2]).strip())
                print("Direccion de Suministro: " + str(txtfromfpdf[4]).strip() + " " + str(txtfromfpdf[5]).strip())

            #print(txtfromfpdf)
    else:
        print("******* Recibo no encontrado del suministro: " + str(i) + " *******")
```
